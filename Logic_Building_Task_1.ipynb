```python
# Problem 1: User Login Check
username = "admin"
password = "1234"

entered_username = input("Enter username: ")
entered_password = input("Enter password: ")

print("Login Successful" if entered_username == username and entered_password == password else "Invalid Credentials")
```

```python
# Problem 2: Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]

pass_cnt = 0
fail_cnt = 0

for m in marks:
    if m >= 50:
        pass_cnt += 1
    else:
        fail_cnt += 1

print("Pass / Fail Analysis")
print(f"Total Students : {len(marks)}")
print(f"Passed         : {pass_cnt}\nFailed         : {fail_cnt}")
```

```python
# Problem 3: Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]

print("Cleaned Names:")
print(cleaned_names)
```

```python
# Problem 4: Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

print("Message Length Analysis:")
for msg in messages:
    length = len(msg)
    if length > 10:
        print(f"'{msg}' → Length: {length} ⚠️ (Too Long)")
    else:
        print(f"'{msg}' → Length: {length} ✅")
```

```python
# Problem 5: Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("Log Analysis:")
print(f"Total Logs   : {len(logs)}")
print(f"Total Errors : {error_count}")
```
